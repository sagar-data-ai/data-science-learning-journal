9. Region

```python
df['region'].value_counts()
```

```python
df['region'].isnull().sum()    #1 missing 
df[df['region'].isnull()][['name', 'state','region']]

#panjeeri north india's famous sweet dish
```

```python
#filling missing region
df.loc[df['name'].str.lower() == 'panjeeri', 'region'] = 'north'
```

```python
df['region'] = df['region'].str.lower()
```

```python
df[df['region'] == '-1']
```

```python
#finding names of dishes with -1 vlaue
missing_region_names = df.loc[df['region'] == '-1', 'name'].tolist()
missing_region_names
```

```python
# mapping for regions
region_mapping = {
    'kaju katli': 'west',
    'kheer': 'north',
    'laddu': 'west',
    'nankhatai': 'west',
    'khichdi': 'west',
    'kulfi falooda': 'north',
    'lauki ki subji': 'north',
    'pani puri': 'west',
    'papad': 'west',
    'samosa': 'north',
    'upma': 'south',
    'brown rice': 'south',
    'red rice': 'south'
}

df['region'] = df.apply(lambda x: region_mapping.get(x['name'].lower(), x['region']), axis=1)
```
